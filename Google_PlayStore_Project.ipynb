# Importing the Librabies

```python
#importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading the dataset
pd.set_option("display.max_columns",None)
df=pd.read_csv("C:\\Users\\Sai Ram\\Desktop\\google playstore project\\googleplaystore.csv")
print(df)
```

```python
#Information About Our Dataset Like Total Number Rows, Total Number of Columns,Datatypes of Each Column And  Memory Requirement
df.info()
```

# Data Cleaning

```python
#remove duplicates
df.drop_duplicates(inplace=True)
print(df)
```

```python
# Drop rows with missing ratings
df.dropna(subset=['Rating'], inplace=True)
print(df)
```

```python
# Convert 'Installs' to numeric
df['Installs'] = df['Installs'].str.replace('+', '', regex=False).str.replace(',', '', regex=False).astype(int)
print(df)
```

```python
# Ensure data cleaning is done
print("Cleaned data:")
print(df.info())
```

# Data Visualisation

```python
# Distribution of Ratings
plt.figure(figsize=(5, 4))
plt.hist(df['Rating'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of App Ratings', fontsize=14)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
type_counts = df['Type'].value_counts()

plt.figure(figsize=(5,5))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', colors=['gold', 'lightblue'], startangle=90, explode=[0.1, 0])
plt.title('Free vs Paid Apps', fontsize=16)
plt.show()
```

```python
# Total installs per category
installs = df.groupby('Category')['Installs'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(5,5))
plt.barh(installs.index, installs.values, color='coral')
plt.title('Top 5 Categories by Total Installs', fontsize=16)
plt.xlabel('Total Installs', fontsize=14)
plt.ylabel('Category', fontsize=14)
plt.gca().invert_yaxis()  # Invert y-axis for better appearance
plt.show()
```

```python
# Cumulative installs
top_installs = df.groupby('Category')['Installs'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(7,6))
plt.fill_between(range(len(top_installs)), top_installs.values, color='orange', alpha=0.5)
plt.plot(top_installs.values, marker='o', color='red')
plt.xticks(range(len(top_installs)), top_installs.index, fontsize=12)
plt.title('Cumulative Installs Across Top Categories', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Total Installs', fontsize=14)
plt.grid(alpha=0.7)
plt.show()
```

```python
# Scatter Plot: Rating vs. Installs
plt.figure(figsize=(5,4))
plt.scatter(df['Installs'], df['Rating'], alpha=0.6, c='green', edgecolor='k')
plt.title('Rating vs. Installs', fontsize=14)
plt.xlabel('Installs', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.xscale('log')  # To better display Install data
plt.grid(alpha=0.5)
plt.show()
```
